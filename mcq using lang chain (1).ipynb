```python
import fitz

doc = fitz.open("/content/budget_speech.pdf")

text = ""

for page in doc:
    text += page.get_text()

#print(text[:1000])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

chunks = splitter.split_text(text)

print(len(chunks))
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    api_key='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',
    model_name='llama-3.3-70b-versatile'
)
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["text"],
    template="""
Generate 5 MCQ questions from the text.

TEXT:
{text}

Format:
Question:
A.
B.
C.
D.
Correct Answer:
"""
)
```

```python
chain = prompt | llm

response = chain.invoke({
    "text": chunks[0]
})

print(response.content)
```
